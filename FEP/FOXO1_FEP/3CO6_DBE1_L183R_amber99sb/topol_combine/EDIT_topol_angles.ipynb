```python

# **************************************************************************************************************

# L2R index is 34. LU1 index is 68. L2R is before LU1
# the 1st atom: N of LU1 in f_mcbp is 1061
# the 1st atom: N of LU1 in f_hybrid is 1069
# 1069 - 1061 = 8

special_atom_list = \
['CX1', 'A2C', 'A3C', 'Y1', 'CX2', 'Y2', 'CX3', 'Y3', 'M1']
special_atom_index_list_mcpb = \
[1063, 1065, 1068, 1079, 1112, 1126, 1162, 1179, 1572]
special_atom_index_list_hybrid = \
[i+8 for i in special_atom_index_list_mcpb]

print(special_atom_index_list_hybrid)
```

```python
##### special atoms in 
# angles_mcpb_list

list_angles_mcpb_special = \
[angles_mcpb_list[i] for i in range(len(angles_mcpb_list)) if (i > 1) and \
(int(angles_mcpb_list[i][0]) in special_atom_index_list_mcpb or \
 int(angles_mcpb_list[i][1]) in special_atom_index_list_mcpb or \
 int(angles_mcpb_list[i][2]) in special_atom_index_list_mcpb)]

print(list_angles_mcpb_special)
print(len(list_angles_mcpb_special))
```

```python

## modify the atom index in list_angles_mcpb_special by "+ 8"
## to the corresponding atom index in list_angles_hybrid_special

for i in list_angles_mcpb_special:
    
    result_1 = int(i[0]) + 8
    result_2 = int(i[1]) + 8
    result_3 = int(i[2]) + 8
    i[0] = result_1
    i[1] = result_2
    i[2] = result_3

print(list_angles_mcpb_special)
print(len(list_angles_mcpb_special))
```

```python
##### special atoms in 
# angles_hybrid_list

list_angles_hybrid_special = \
[angles_hybrid_list[i] for i in range(len(angles_hybrid_list)) if i > 1 and \
(int(angles_hybrid_list[i][0]) in special_atom_index_list_hybrid or \
 int(angles_hybrid_list[i][1]) in special_atom_index_list_hybrid or \
 int(angles_hybrid_list[i][2]) in special_atom_index_list_hybrid)]

print(list_angles_hybrid_special)
print(len(list_angles_hybrid_special))
```

```python

## compare modified list_angles_mcpb_special and 
## list_angles_hybrid_special

count_match = 0

# Here range is the number for angles
for i in range(63):
    
    atom1_index = int(list_angles_mcpb_special[i][0])
    atom2_index = int(list_angles_mcpb_special[i][1])
    atom3_index = int(list_angles_mcpb_special[i][2])
    
    # Here range is the number for angles
    for j in range(63):
        
        if (atom1_index == int(list_angles_hybrid_special[j][0])) and \
           (atom2_index == int(list_angles_hybrid_special[j][1])) and \
           (atom3_index == int(list_angles_hybrid_special[j][2])):
            count_match += 1
            print(str(atom1_index) + " " + str(atom2_index) + " " + str(atom3_index))
            break

print(count_match)

## *************************************************************************************************************

new_list = []

# Here range is the number for angles
for x in range(63):
    
    atom1_index = int(list_angles_mcpb_special[x][0])
    atom2_index = int(list_angles_mcpb_special[x][1])
    atom3_index = int(list_angles_mcpb_special[x][2])
    
    for y in range(63):
        
        if (atom1_index == int(list_angles_hybrid_special[y][0])) and \
           (atom2_index == int(list_angles_hybrid_special[y][1])) and \
           (atom3_index == int(list_angles_hybrid_special[y][2])):
            
            new_angle_line = [atom1_index, atom2_index, atom3_index, int(list_angles_mcpb_special[x][3]), 
                             float(list_angles_mcpb_special[x][4]), float(list_angles_mcpb_special[x][5])]
            
            new_list.append(new_angle_line)
            
            break

print(new_list)
print(len(new_list))
```
