```python
bonds_mcpb_list = []

while True:
    
    f_mcpb_line = f_mcpb.readline()
    ## strip() makes space line become empty line   
    f_mcpb_listPerLine = f_mcpb_line.strip().split()    
    if len(f_mcpb_listPerLine) <= 2:
        continue
        
    if f_mcpb_listPerLine[1] == "bonds":
        bonds_mcpb_list.append(f_mcpb_listPerLine)
        break

while True:
    
    f_mcpb_line = f_mcpb.readline()
    f_mcpb_listPerLine = f_mcpb_line.strip().split()
    if len(f_mcpb_listPerLine) <= 2: 
        break        
    bonds_mcpb_list.append(f_mcpb_listPerLine)    

        
f_mcpb.close()

```

```python

# **************************************************************************************************************

# L2R index is 34. LU1 index is 68. L2R is before LU1
# the 1st atom: N of LU1 in f_mcbp is 1061
# the 1st atom: N of LU1 in f_hybrid is 1069
# 1069 - 1061 = 8

special_atom_list = \
['CX1', 'A2C', 'A3C', 'Y1', 'CX2', 'Y2', 'CX3', 'Y3', 'M1']
special_atom_index_list_mcpb = \
[1063, 1065, 1068, 1079, 1112, 1126, 1162, 1179, 1572]
special_atom_index_list_hybrid = \
[i+8 for i in special_atom_index_list_mcpb]

print(special_atom_index_list_hybrid)
```

```python
##### special atoms in 
# bonds_mcpb_list

list_bonds_mcpb_special = \
[bonds_mcpb_list[i] for i in range(len(bonds_mcpb_list)) if i > 1 and \
(int(bonds_mcpb_list[i][0]) in special_atom_index_list_mcpb or \
 int(bonds_mcpb_list[i][1]) in special_atom_index_list_mcpb)]

print(list_bonds_mcpb_special)
print(len(list_bonds_mcpb_special))
```

```python

## modify the atom index in list_bonds_mcpb_special by "+ 8"
## to the corresponding atom index in list_bonds_hybrid_special

for i in list_bonds_mcpb_special:
    
#     print(int(i[0]))
#     print(int(i[1]))
#     break
    
    result_1 = int(i[0]) + 8
    result_2 = int(i[1]) + 8
    i[0] = result_1
    i[1] = result_2

print(list_bonds_mcpb_special)
print(len(list_bonds_mcpb_special))
```

```python
##### special atoms in 
# bonds_hybrid_list

list_bonds_hybrid_special = \
[bonds_hybrid_list[i] for i in range(len(bonds_hybrid_list)) if i > 1 and \
(int(bonds_hybrid_list[i][0]) in special_atom_index_list_hybrid or \
 int(bonds_hybrid_list[i][1]) in special_atom_index_list_hybrid)]

print(list_bonds_hybrid_special)
print(len(list_bonds_hybrid_special))
```

```python

## compare modified list_bonds_mcpb_special and 
## list_bonds_hybrid_special

count_match = 0

for i in range(24):
    
    atom1_index = int(list_bonds_mcpb_special[i][0])
    atom2_index = int(list_bonds_mcpb_special[i][1])
    
    for j in range(24):
        
        if (atom1_index == int(list_bonds_hybrid_special[j][0])) and \
           (atom2_index == int(list_bonds_hybrid_special[j][1])):
            count_match += 1
            print(str(atom1_index) + " " + str(atom2_index))
            break

print(count_match)

## *************************************************************************************************************

new_list = []

for x in range(24):
    
    atom1_index = int(list_bonds_mcpb_special[x][0])
    atom2_index = int(list_bonds_mcpb_special[x][1])
    
    for y in range(24):
        
        if (atom1_index == int(list_bonds_hybrid_special[y][0])) and \
           (atom2_index == int(list_bonds_hybrid_special[y][1])):
            
            new_bond_line = [atom1_index, atom2_index, int(list_bonds_mcpb_special[x][2]), 
                             float(list_bonds_mcpb_special[x][3]), float(list_bonds_mcpb_special[x][4])]
            
            new_list.append(new_bond_line)
            
            break

print(new_list)
print(len(new_list))
```

```python

# replace special atoms forming bonds in hybrid with corresponding bonds in mcpb

print(len(bonds_hybrid_list))
print(bonds_hybrid_list)

new_bonds_hybrid_list = []

f_bonds = open("/home/leiqian/Documents/Tutorial/tutorial_fep/foxo1_fep/fep-foxo1-with-mg/new_bonds.itp", "w")

for i in range(2, 1602, 1):
    
    atom1_index = int(bonds_hybrid_list[i][0])
    atom2_index = int(bonds_hybrid_list[i][1])
    
    match = 0
    
    for j in range(24):
        if (atom1_index == new_list[j][0]) and \
           (atom2_index == new_list[j][1]):
            
            new_list[j].append(";<<<<<<<<< update bonds")
            new_bonds_hybrid_list.append(new_list[j])
            new_str1 = str(new_list[j]).replace('\'', '').replace('[', '').replace(']', '').replace(',', '    ')
            f_bonds.write(new_str1 + "\n")
            match += 1
            break
      
    if match == 0:
        new_bonds_hybrid_list.append(bonds_hybrid_list[i])
        new_str2 = str(bonds_hybrid_list[i]).replace('\'', '').replace('[', '').replace(']', '').replace(',', '    ')
        f_bonds.write(new_str2 + "\n")

        
print(new_bonds_hybrid_list)                        
print(len(new_bonds_hybrid_list))

f_bonds.close()

```
